```
import matplotlib.pyplot as plt
import numpy as np

e = 2.718281827

#N contains the values of n and val contains corresponding values of the expression
N = []
val = []

#min and max value of n and step
l = 100
r = 10000
step = 100

for i in range(l, r+1, step):
    N.append(i)
    value = (1 + 1/i) ** i
    val.append(value)

E = [e]*(len(N))
plt.plot(N, val) #plots the function y = (1 + 1/x)^x
plt.plot(N, E, color = 'red') # plots y = e
```
